```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sleep.csv")
```

```python
df
```

```python
fig, ax, = plt.subplots()
ax.barh(df['Age group'].loc[::-1],df['Recommended lower'].loc[::-1],height = 0.6,color='#4789be',edgecolor='white')
ax.barh(df['Age group'].loc[::-1],df['Recommended upper'].loc[::-1],height = 0.3,color = '#051b2c',align='edge', edgecolor='white') 
ax.set_xlabel('Hours')
ax.set_xlim(0,18)
ax.text(0.44, 1.02, 'Recommended lower limit', ha='right',  transform=ax.transAxes, fontsize=10, color='#4789be')
ax.text(0.57, 1.02, 'Recommended upper limit', ha='left', transform=ax.transAxes, fontsize=10,color = '#051b2c')
ax.set_title("National Sleep Foundation's recommended hours of sleep",fontsize=13.5,pad=20)
fig.dpi = 300
```

```python
fig, ax = plt.subplots()

ax.scatter(df['Recommended lower'].loc[::-1],df['Age group'].loc[::-1],color = '#051b2c',edgecolors='black')
ax.scatter(df['Recommended upper'].loc[::-1],df['Age group'].loc[::-1],color='#4789be',edgecolors='black')
ax.hlines(y=df['Age group'].loc[::-1],xmin=df['Recommended lower'].loc[::-1], xmax=df['Recommended upper'].loc[::-1], colors='black',zorder=0)

ax.set_title("National Sleep Foundation's recommended hours of sleep",fontsize=13.25,pad=30)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax.set_ylim(-0.5,8+0.5)
ax.set_xticks(ticks=[0,6,8,10,12,14,18,24])
ax.set_yticklabels(df['Age group'].loc[::-1])
ax.tick_params(axis='y', length=0)
ax.xaxis.tick_top()
fig.dpi = 200
```
